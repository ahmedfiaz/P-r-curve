```python
T=f1.variables['T'][9:,:126,:,:]
q=f1.variables['q'][9:,:126,:,:]
lev=f1.variables['lev'][8:]

dp=np.diff(lev)
pw=np.sum(q*dp[:,None,None,None],axis=0)
print T.shape,conv_rain.shape
```

## Saturation vapor pressure

```python
a=np.array([6.11176750,0.443986062,0.143053301e-1,0.265027242e-3,0.302246994e-5,0.203886313e-7,0.638780966e-10])
Tr=273.15
es=a[0]+a[1]*(T-Tr)+a[2]*(T-Tr)**2+a[3]*(T-Tr)**3+a[4]*(T-Tr)**4+a[5]*(T-Tr)**5+a[6]*(T-Tr)**6
```

```python
epsilon=0.622
qs=es*epsilon/(lev[1:,None,None,None]-(1-epsilon)*es) #saturation specific humidity
pw_sat=np.sum(qs*dp[:,None,None,None],axis=0)
```

```python
crh=pw/pw_sat
print crh.min()
```
