```python
import numpy as np
import ncdump
from netCDF4 import Dataset
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from mpl_toolkits.basemap import Basemap,cm
%matplotlib inline
import matplotlib.patches as patches
from matplotlib.pylab import *

```

```python
f=Dataset("/Users/fiaz_ahmed/python/org_precip/trmm_rain_199802.nc","r")
a,b,c=ncdump.ncdump(f)
lat1=f.variables['lat'][:]
lon1=f.variables['lon'][:]

lat1=np.arange(-33.75,36.25,2.5)
lon1=np.arange(-178.75,181.25,2.5)
lon1[lon1<0]+=360
ind=np.argsort(lon1)
lon1=lon1[ind]

conv_rain=f.variables['conv_rain'][:,:].T*24
strat_rain=f.variables['strat_rain'][:,:].T*24
shallow_rain=f.variables['shallow_rain'][:,:].T*24
shallow_rain[shallow_rain>1e10]=0
print shallow_rain.max()

conv_rain=conv_rain[:,ind]
strat_rain=strat_rain[:,ind]
shallow_rain=shallow_rain[:,ind]
```

```python
fig=plt.figure(figsize=(8,12))
ax = fig.add_axes([0.05, 0.05, 0.95, 0.95])
#shallow_rain=np.ma.masked_where(shallow_rain == 0,shallow_rain)
m=Basemap(llcrnrlon=1.25,llcrnrlat=-35,urcrnrlon=358.75,urcrnrlat=35,projection='mill',ax=ax)
m.drawcoastlines(linewidth=0.25)
m.drawcountries(linewidth=0.25)
clevs=np.arange(0,2,0.05)
lons, lats = np.meshgrid(lon1, lat1)
x,y=m(lons,lats)
cbaxes = fig.add_axes([.25, 0.425, 0.5, 0.0125]) 
cs = m.contourf(x,y,shallow_rain,levels=clevs,cmap=plt.get_cmap('RdYlBu_r'),extend='both')
CB1 = plt.colorbar(cs, orientation='horizontal',label='mm/hr',cax=cbaxes)
CB1.ax.tick_params(labelsize=14)
ax.set_title('Shallow Rain')
```
