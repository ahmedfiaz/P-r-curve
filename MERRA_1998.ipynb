```python
import numpy as np
import ncdump
from netCDF4 import Dataset
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from mpl_toolkits.basemap import Basemap,cm
%matplotlib inline
import matplotlib.patches as patches
from matplotlib.pylab import *
from scipy.interpolate import RectSphereBivariateSpline

```

```python
f=Dataset('/Users/fiaz_ahmed/python/org_precip/merra_daily_Tq_1998.nc','r')
ftrmm=Dataset('/Users/fiaz_ahmed/python/org_precip/trmm_precip_daily_1998.nc','r')

a,b,c=ncdump.ncdump(ftrmm)

lat1=f.variables['lat'][:]
lon1=f.variables['lon'][:]
lev=f.variables['lev'][:]/100

lat2=ftrmm.variables['lat'][:]
lon2=ftrmm.variables['lon'][:]
lon2[lon2<0]+=360

ind= np.argsort(lon2)
print lon2[ind]
print type(ind)

print ind
dp=np.diff(lev)
print dp

T=f.variables['T'][:]
q=f.variables['q'][:]

large_precip=f.variables['large_precip'][:]
conv_precip=f.variables['conv_precip'][:]

```

```python
conv_rain=ftrmm.variables['conv_rain'][:,:,:]
strat_rain=ftrmm.variables['strat_rain'][:,:,:]
shallow_rain=ftrmm.variables['shallow_rain'][:,:,:]
total_rain=ftrmm.variables['total_rain'][:]

conv_rain=conv_rain[:,:,ind]
strat_rain=strat_rain[:,:,ind]
shallow_rain=shallow_rain[:,:,ind]
total_rain=total_rain[:,:,ind]

print conv_rain.shape


```

```python
Tr=273.15
T=np.float64(f.variables['T'][1:,:,:,:])
q=f.variables['q'][1:,:,:,:]
pw=np.sum(q*dp[:,None,None,None]*-1,axis=0)
print pw.max()
```

```python
a=np.array([6.11176750,0.443986062,0.143053301e-1,0.265027242e-3,0.302246994e-5,0.203886313e-7,0.638780966e-10])
Tr=273.15
es=a[0]+a[1]*(T-Tr)+a[2]*(T-Tr)**2+a[3]*(T-Tr)**3+a[4]*(T-Tr)**4+a[5]*(T-Tr)**5+a[6]*(T-Tr)**6
```

```python
epsilon=0.622
qs=es*epsilon/(lev[1:,None,None,None]-(1-epsilon)*es) #saturation specific humidity
pw_sat=np.sum(qs*dp[:,None,None,None]*-1,axis=0)
```

```python
crh=pw/pw_sat
crh_io=crh[:,6:22,67:91]

lp=large_precip[:,6:22,67:91]
cp=conv_precip[:,6:22,67:91]

trmm_conv=conv_rain[:,6:22,67:91]/24
trmm_strat=strat_rain[:,6:22,67:91]/24
trmm_shallow=shallow_rain[:,6:22,67:91]/24
trmm_tot=total_rain[:,6:22,67:91]/24

trmm_shallow[trmm_shallow>1e10]=0.0
trmm_tot=trmm_conv+trmm_strat+trmm_shallow

print crh_io.min()
print lp.min()
```

```python
fig,axx=plt.subplots(2,2,figsize=(10,10))
ax=axx[1,0]
ax.scatter(crh_io,trmm_shallow)
ax.set_ylim(0,20)
ax.set_title('Shallow Precipitation')
ax.set_ylabel('mm/hr')

ax=axx[0,0]
ax.scatter(crh_io,trmm_tot)
ax.set_ylim(0,20)
ax.set_title('Total Precipitation')
ax.set_ylabel('mm/hr')


ax=axx[0,1]
ax.scatter(crh_io,trmm_conv)
ax.set_ylim(0,20)
ax.set_title('Convective Precipitation')
ax.set_ylabel('mm/hr')


ax=axx[1,1]
ax.scatter(crh_io,trmm_strat)
ax.set_ylim(0,20)
ax.set_title('Stratiform Precipitation')
ax.set_ylabel('mm/hr')

plt.tight_layout()
plt.savefig('TRMM_precip_Moisture.png',bbox_inches='tight')

```

```python
crh_bins=np.arange(0,1.01,.01)

crain_bins=np.zeros((crh_bins.size-1))
srain_bins=np.zeros_like(crain_bins)
train_bins=np.zeros((crh_bins.size-1))

for i,j in enumerate(crh_bins):
    if i <crh_bins.size-1:
        c1,c2=crh_bins[i],crh_bins[i+1]
        crain_bins[i]=trmm_conv[(crh_io>=c1) & (crh_io<c2)].mean()
        train_bins[i]=trmm_tot[(crh_io>=c1) & (crh_io<c2)].mean()
        srain_bins[i]=trmm_strat[(crh_io>=c1) & (crh_io<c2)].mean()
        
print crain_bins
```

```python
fig,axx=plt.subplots(3,1,figsize=(10,10))

ax=axx[0]
ax.scatter(crh_bins[:-1],train_bins)
ax.set_ylim(0,4)
ax.set_title('Total Precipitation')
ax.set_ylabel('mm/hr')

ax=axx[1]
ax.scatter(crh_bins[:-1],crain_bins)
ax.set_ylim(0,4)
ax.set_title('Convective Precipitation')
ax.set_ylabel('mm/hr')


ax=axx[2]
ax.scatter(crh_bins[:-1],srain_bins)
ax.set_ylim(0,4)
ax.set_title('Stratiform Precipitation')
ax.set_ylabel('mm/hr')

plt.tight_layout()
plt.savefig('TRMM_Histograms.png')

```
