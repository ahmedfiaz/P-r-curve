```python
fr=Dataset('/Users/fiaz_ahmed/python/org_precip/isccp_daily_1998_2008_Tq.nc','r')
lat=fr.variables['lat'][:]
lon=fr.variables['lon'][:]
print lon

conv_rain_ave=fr.variables['conv_rain'][:]/24
strat_rain_ave=fr.variables['strat_rain'][:]/24
shallow_rain_ave=fr.variables['shallow_rain'][:]/24
total_rain_ave=conv_rain_ave+shallow_rain_ave+strat_rain_ave

print strat_rain_ave.shape
```

```python
print conv_rain_ave.max(),strat_rain_ave.max(),strat_rain_ave.min(),shallow_rain_ave.max(),shallow_rain_ave.min()

sfrac=strat_rain_ave/(strat_rain_ave+conv_rain_ave+shallow_rain_ave)
```

```python
fig=plt.figure(figsize=(11,12))
ax = fig.add_axes([0.05, 0.05, 0.95, 0.3])

m=Basemap(llcrnrlon=1.25,llcrnrlat=-35,urcrnrlon=358.75,urcrnrlat=35,projection='mill',ax=ax)
m.drawcoastlines(linewidth=0.25)
m.drawcountries(linewidth=0.25)
clevs=np.arange(0,0.5,.025)
lons, lats = np.meshgrid(lon, lat)
x,y=m(lons,lats)
#cbaxes = fig.add_axes([.25, 0.325, 0.5, 0.0125]) 
cs = m.contourf(x,y,np.nanmean(sfrac[:,:,:],0),levels=clevs,cmap=plt.get_cmap('PuBu'),extend='both')
ax.set_title('Stratiform')
```
