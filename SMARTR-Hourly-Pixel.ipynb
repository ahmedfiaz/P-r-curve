## Logarithmic fit function

```python
def log_fit(x,y,thresh):
    
    from scipy import stats
    
    ''' 
        Given two one-dimensional arrays x and y
        fit a non-linear curve, by taking the logarithm of the non-linear function, for data points x and y.
        thresh indicates the rain threshold to use to accomplish the logarithm fit

    '''
    print x.shape,y.shape
    lx,ly=np.log(x),np.log(y)
    mask = np.isfinite(lx) & np.isfinite(ly) & (x>=thresh)
    slope, intercept, r_value, p_value, std_err = stats.linregress(lx[mask],ly[mask])
    
    return np.exp(intercept), slope, r_value, p_value, std_err
    
```
